# _Concatenat All csv Files togather_

### concat on basis of axis 1 means side by side

```python
df_concat = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11 ,df_12, df_13, df_14, df_15], axis=1)
```

```python
df_concat
```

### concatenate on axis 0 but it will not removing indexing

```python
df_concat = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11 ,df_12, df_13, df_14, df_15], axis=0)
```
